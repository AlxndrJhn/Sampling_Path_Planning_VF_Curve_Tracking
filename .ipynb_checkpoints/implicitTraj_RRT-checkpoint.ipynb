# Simple UAV 3D Model
Taken from Gon√ßalvez et al. 2011
See (3) for the state derivative description ($\theta_i$ is in the text paragraph below (3))

```python
%matplotlib inline
#import pdb
import numpy as np
#import matplotlib
#import matplotlib.pyplot as plt

inputBounds = dict()
inputBounds["vthetamin"] = -np.pi/5
inputBounds["vthetamax"] =  np.pi/5
inputBounds["vzmin"] = -0.1
inputBounds["vzmax"] =  0.1
inputBounds["vfmin"] = 1 # minimum forward velocity
inputBounds["vfmax"] = 2

# time derivative of the state vector
def dq(q,u):
    # unpack the state vector
    x = q[0]
    y = q[1]
    z = q[2]
    theta = q[3]

    # unpack the input vector
    vtheta = u[0]
    vz = u[1]
    vf = u[2]
    
    # check the minimum forward velocity
    if vf < inputBounds["vfmin"] or vf > inputBounds["vfmax"]:
        raise ValueError('The input forward velocity is out of bound')
    if vz < inputBounds["vzmin"] or vz > inputBounds["vzmax"]:
        raise ValueError('The input z velocity is out of bound')
    if vtheta < inputBounds["vthetamin"] or vtheta > inputBounds["vthetamax"]:
        raise ValueError('The input angular velocity is out of bound')
        
    # compute deltas
    dx = vf*np.cos(theta)
    dy = vf*np.sin(theta)
    dz = vz
    dtheta = vtheta
    
    # state time derivative
    dq = np.zeros(4)
    dq[0] = dx
    dq[1] = dy
    dq[2] = dz
    dq[3] = dtheta
    
    # return the state time derivative
    return dq
```

# The Vector-Field VF(x) #
(taken from Goncalves2011, see (1) and (2) for the definitions of $\alpha_i$ )

```python
import numpy as np

# super simple VF wih a limit cycle in a plane ring in he 3D space

# center of the ring that will span in the xy-plane with unifom heigt zc
xc = 0
yc = 0
zc = 1

# radius
r = 1

# weights
wc = 1
watt = 1

#Vector Field function as R¬≥->R¬≥
def VF(pos):
    x = pos[0]
    y = pos[1]
    z = pos[2]

    alpha = np.arctan2(y-yc,x-xc)

    # circulating term, unit vector perpencular to the circle
    circ = np.zeros(3)
    circ[0] = -np.sin(alpha)
    circ[1] = np.cos(alpha)

    # attracting term, unit vector towards the ring
    xt = xc-x
    yt = yc-y
    zt = zc-z
    l = np.linalg.norm([xt,yt])

    att = (l-r)/l*np.array([xt,yt,0])
    att[2] = zt
    

    # weighting
    vec = wc*circ+watt*att
    
    return vec/np.linalg.norm(vec)

# just show some integral curve
from scipy.integrate import odeint
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt


pos = [3, 3, 3]

def f(pos,t0):
    return VF(pos)*0.1

curve = odeint(func=f,y0=pos,t=range(0,400))


mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(curve[:,0], curve[:,1], curve[:,2], label='integral curve')
ax.legend()

plt.show()
```

# iCRRT Algorithm #

```python
# Tree Data Structure
class Tree(object):
    def __init__(self, data):
        self.data = list([data])
        self.parent = list([-1])
        self.edges = list([-1])
        self.NumNodes = 1

    def add_child(self, parent, data, edge):
        self.data.append(data)
        self.parent.append(parent)
        self.edges.append(edge)
        self.NumNodes += 1
        
    def get_children(self,node):
        return [i for i, j in enumerate(self.parent) if j == node]
    
    def get_data(self,node):
        return self.data[node]
    
    def plot(self):
        #pdb.set_trace()
        import numpy as np
        import matplotlib.pyplot as plt
        from mpl_toolkits.mplot3d import Axes3D
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        
        
        for i in range(0,self.NumNodes):
            s = self.get_data(i)
            for j in self.get_children(i):
                e = self.get_data(j)
                ax.plot([s[0], e[0]], [s[1],e[1]],zs=[s[2],e[2]],color='black')
        root = self.get_data(0)
        ax.scatter(root[0],root[1],root[2],c='red', marker='o')
        
        
```

```python
import numpy as np
import random as rnd

# Environment is a 10x10x10 3D cube
bounds = dict()
bounds["xmin"] = -5
bounds["xmax"] =  5
bounds["ymin"] = -5
bounds["ymax"] =  5
bounds["zmin"] = -5
bounds["zmax"] =  5

# create some obstacles
def CreateSphericalObstacles(envBounds=bounds,N=10,r=0.5):
    import numpy.random as rnd
    
    Obstacles = list()
    for i in range(0,N):
        ob = dict()
        pos = np.array([0.0,0.0,0.0])
        pos[0] = rnd.uniform(envBounds["xmin"],envBounds["xmax"])
        pos[1] = rnd.uniform(envBounds["ymin"],envBounds["ymax"])
        pos[2] = rnd.uniform(envBounds["zmin"],envBounds["zmax"])
        ob["pos"] = pos
        ob["r"] = r
        Obstacles.append(ob)
    
    return Obstacles
Obstacles = CreateSphericalObstacles()

# distance measure, weightened sum of difference in position and angle
# np.sqrt(300) is the diagonal in the R¬≥ space with edge length 10
def pseudoDist(q1,q2,divEuclDist=np.sqrt(300),angularDiffWeight=0.3):
    # q1 is the start and q2 is the final point, only matters for the sign of the angular difference
    euclDist = np.linalg.norm(q2[0:3]-q1[0:3])
    anglDist = np.arctan2(np.sin(q2[3]-q1[3]), np.cos(q2[3]-q1[3]))
    
    # normalizes both components to 0..1 interval, weights the angular component
    pseudo = np.sqrt(np.power(euclDist/divEuclDist,2)+np.power(angularDiffWeight*anglDist/np.pi,2))
    
    return pseudo,euclDist,anglDist
    
# random Sample
def rndSample(envBounds=bounds, mustBeFree=False):
    q = np.array([0.0,0.0,0.0,0.0])
    while not q.any() or (mustBeFree and not isBlocked(q)):
        q[0] = rnd.uniform(bounds["xmin"],bounds["xmax"])
        q[1] = rnd.uniform(bounds["ymin"],bounds["ymax"])
        q[2] = rnd.uniform(bounds["zmin"],bounds["zmax"])
        q[3] = rnd.uniform(-np.pi,np.pi)
    
    return q

# collision check if q is inside one of the obstacles
def isBlocked(q,Obs=Obstacles):
    for o in Obs:
        if np.linalg.norm(o["pos"]-q[0:3])<=o["r"]:
            return True
    
    return False

# find nearest node on the tree, returns ID of qnear, data of qnear, distance
def nearestNode(Tree,qrnd,metric=pseudoDist):
    minDist,_,_ = metric(Tree.get_data(0),qrnd)
    minNode = Tree.get_data(0)
    minNodeID = 0
    for IDchild in range(1,Tree.NumNodes):
        dist,_,_ = metric(Tree.get_data(IDchild),qrnd)
        if dist < minDist:
            minNode = Tree.get_data(IDchild)
            minNodeID = IDchild
            minDist = dist
    
    return minNodeID, minNode, minDist
        
        
    
# try to steer from node q1 to node q2 and returns qnew, ubest
def steer(q1,q2):
    from scipy.integrate import odeint
    
    # Parameters, adjust TODO
    timeStep = 0.1 # qnew is timeStep seconds away from q1 
    vfsteps = 5
    vthetasteps = 5
    
    # calculate the best vz, might be bounded
    dz = q2[2]-q1[2]
    if dz>=0:
        if timeStep*inputBounds["vzmax"]<dz:
            vz = inputBounds["vzmax"]
        else:
            vz = dz/timeStep
    else:
        if timeStep*inputBounds["vzmin"]>dz:
            vz = inputBounds["vzmin"]
        else:
            vz = dz/timeStep
    # sweep the input space for vtheta and vf to find best steering, static steps
    ubest = np.zeros(3)
    distbest = np.Inf
    qbest = np.zeros(4)
    for vf in np.linspace(inputBounds["vfmin"],inputBounds["vfmax"],vfsteps):
        for vtheta in np.linspace(inputBounds["vthetamin"],inputBounds["vthetamax"],vthetasteps):
            utmp = np.array([vtheta,vz,vf])
            f = lambda q,t0: dq(q,utmp)
            qtmp = odeint(func=f,y0=q1,t=[0,timeStep])
            disttmp,_,_ = pseudoDist(qtmp[1],q2)
            if disttmp<distbest:
                ditbest = disttmp
                ubest = utmp
                qbest = qtmp[1]
    return qbest, ubest

# extend Tree at node qnear with qnew
def extend(Tree,IDqnear,qnew,edge):
    Tree.add_child(IDqnear,qnew,edge)
```

# Experiment

```python
import pdb

# Experiment
chanceForVF = 0.05 # 5% chance to follow directly the VF from a chosen qnear (based on a random sample)
MaxNodes = 2000

qstart = np.array([0.,0.,0.,0.])

T = Tree(qstart)

while T.NumNodes<MaxNodes:
    # random sample qrnd
    qrnd = rndSample()
    
    # get qnear
    qnearID, qnear, dist = nearestNode(Tree=T,qrnd=qrnd)
    
    # steer from qnear to qrnd -> qnew
    qnew,u = steer(qnear,qrnd)
    
    # check if qnew is blocked
    if not isBlocked(qnew):
        # if no, add qnew
        extend(T,qnearID,qnew,u)
    
T.plot()
```

# Implicit Trajectory combined with RRT #
- Work space: 3D with obstacles
- Obstacle type: static and known
- Agent Model: Dubins airplane (Owen2015)
- Integration Algorithm: 4th Order Runge Kutta
- Implicit Curve: Time-invariant Circle
- Distance Metric: admissionable heuristic, based on time-optimal dubins-airplane paths (altitude difference is integrated in a linear fashion)

### Find blocked Nodes

```python
# Spherical Obstacle Definition
ob = dict()
ob["pos"] = np.array([xc,yc+r,zc])
ob["r"] = 0.2

Obstacles = list()
Obstacles.append(ob)

# Split the qList in subsets whenever an obstacle blocks a node
oldResult = isBlocked(qList[0],Obstacles)
toConnect = []
for i in range(1,len(qList)):
    newResult = isBlocked(qList[i],Obstacles)
    if newResult and not oldResult:
        ilast = i-1
    if not newResult and oldResult:
        ifirst = i
        toConnect.append((ilast,ifirst))
    oldResult = newResult

print(toConnect)

# starting and final configuration
l,f=toConnect[0]
offset = 2
qs = qList[l-offset] # that number is arbitrary
qf = qList[f+offset] # that number is arbitrary

# sample bounds
bounds = dict()
edge = 1.0
bounds["xmin"] = np.min([qs[0],qf[0]])-edge
bounds["xmax"] = np.max([qs[0],qf[0]])+edge
bounds["ymin"] = np.min([qs[1],qf[1]])-edge
bounds["ymax"] = np.max([qs[1],qf[1]])+edge
bounds["zmin"] = np.min([qs[2],qf[2]])-edge
bounds["zmax"] = np.max([qs[2],qf[2]])+edge
```
